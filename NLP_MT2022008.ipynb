```python
def get_tweets(search_query, num_tweets):
    query = search_query
    filters = 'exclude:retweets lang:en -filter:links -filter:replies'          ## collects only english language tweets and exclude retweets, links and replies
    search_term = f'{query} {filters}'
    tweets = set()                                                               # Use a set to store unique tweets
    for i, tweet in tqdm(enumerate(sntwitter.TwitterSearchScraper(search_term).get_items()), total=num_tweets):
        if i >= num_tweets:
            break
        if langdetect.detect(tweet.rawContent) == 'en':
            tweets.add(tweet.rawContent)
    return "".join(list(tweets))                                                         # Convert the set to a list before returning
```

# **Pre-Processing of data**

Using Tweet Preprocessor module of Preprocessor which is a preprocessing library for tweet data written in Python.This library makes it easy to clean, parse or tokenize the tweets.Following has been done for preprocessing :
>

*   removal of urls
*   removal of mentions
*   removal of hashtags
*   removal of emojis
*   normalization of tweets (lowering and punctuation removal)
---


```python
## importing modules for preprocessing
import re
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import TweetTokenizer
from nltk.tokenize import word_tokenize
from nltk.tokenize import word_tokenize
lt = WordNetLemmatizer()
tk = TweetTokenizer()
```

```python
## importing Preprocessor library for basic cleaning of tweet data.
import preprocessor as p
```

```python
## function for removing emojis
def remove_emoji(tweet):
  """Takes a string and removes emojis from it"""
  tweet = tweet.encode('ascii', 'ignore').decode('ascii')
  return tweet
```

```python
## removing links
def remove_links(tweet):
    """Takes a string and removes web links from it"""
    tweet = re.sub(r'http\S+', '', tweet)   # remove http links
    tweet = re.sub(r'bit.ly/\S+', '', tweet)  # remove bitly links
    tweet = tweet.strip('[link]')   # remove [links]
    tweet = re.sub(r'pic.twitter\S+','', tweet)
    return tweet
```

```python
## function for removing hashtags and mentions
def remove_hashtags(tweet):
    """Takes a string and removes any hash tags and mentions from it"""
    # tweet = re.sub('(#[A-Za-z]+[A-Za-z0-9-_]+)', '', tweet)  # remove hash tags
    tweet = re.sub('(@[A-Za-z]+[A-Za-z0-9-_]+)', '', tweet)  # remove mentions
    return tweet
```

```python
## functions for removing audio/video tags or labels
def remove_av(tweet):
    """Takes a string and removes AUDIO/VIDEO tags or labels"""
    tweet = re.sub('VIDEO:', '', tweet)  # remove 'VIDEO:' from start of tweet
    tweet = re.sub('AUDIO:', '', tweet)  # remove 'AUDIO:' from start of tweet
    return tweet
```

```python
## function for tokenizer. We are using "Tweet Tokenizer from NLTK library"
def tokenizer(tweet):    ## function for tokenizing
   """Takes a string and tokenize it using twitter tokenizer from nltk library"""
   tweet = tk.tokenize(tweet)     ## using TweetTokenizer from NLTK library
   return tweet
```

```python
## function for lemmatization. We are using WordNet Lemmatizer
def lemmatizer(tweet):              ## lemmatizing
    """Takes a sentence and lemmatize each word in it"""
    tweet=[lt.lemmatize(word) for word in tweet]
    tweet=tweet[:]
    return tweet
```

```python
## define a string of punctuation symbols
punctuation = '$%&\()*+-/:;<=>[\\]^_`{|}~•@'
```

```python
## function for tweet cleaning which contains all the steps for cleaning tweets.
def pre_processing(data):
  tweet = data['tweets']
  tweet = p.clean(tweet)
  tweet = remove_links(tweet)
  tweet = remove_hashtags(tweet)
  tweet = remove_av(tweet)
  tweet = tweet.lower()                  ## normalizing the tweets by conveting them to lowercase
  tweet = re.sub('[' + punctuation + ']+', ' ', tweet)  # striping punctuation
  tweet = re.sub('\s+', ' ', tweet)  # removing double spacing
  tweet = remove_emoji(tweet)            ## removing emojis
  return tweet
```

```python
## pre-processing the dataset
df['tweets'] = df.apply(pre_processing, axis = 1)
```

```python
## creating embeddings
def convert_tweet_to_feature(text):
  input_data = tokenizer(text['reduced_tweets_700'], max_length = 1200, truncation = True)
  with tokenizer.as_target_tokenizer():
    target_data = tokenizer(text['Auto_summary'], max_length = 150, truncation = True)
  return {
      'input_ids' : input_data['input_ids'],
      'attention_mask' : input_data['attention_mask'],
      'labels' : target_data['input_ids']
  }


data_pt = data.apply(convert_tweet_to_feature, axis = 1)
```

```python
## splitting dataset into train, test and validate
train_dataset = data_pt.loc[0:30].copy()
validate_dataset = data_pt.loc[31:40].copy()
test_dataset = data_pt.loc[41:].copy()
print(len(train_dataset))
print(len(validate_dataset))
print(len(test_dataset))
```

```python
from transformers import DataCollatorForSeq2Seq

seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer, model = model_pegasus_large)
```

```python
## defining the training arguments
from transformers import TrainingArguments, Trainer

trainer_args = TrainingArguments(
    output_dir = 'pegasus-large',
    num_train_epochs = 1,
    warmup_steps = 500,
    per_device_train_batch_size = 1,
    per_device_eval_batch_size = 1,
    weight_decay = 0.05,
    learning_rate = 1e-5,
    logging_steps = 10,
    evaluation_strategy = 'steps',
    eval_steps = 500,
    save_steps = 1e6,
    gradient_accumulation_steps = 16
)
```

```python
## defining the trainer
trainer = Trainer(
    model = model_pegasus_large,
    args = trainer_args,
    tokenizer = tokenizer,
    data_collator = seq2seq_data_collator,
    train_dataset = train_dataset,
    eval_dataset = validate_dataset
)
```

```python
## training the model
trainer.train()
```

```python
rouge1, rouge2, rougeL = 0, 0, 0

def calculateRougeScore(row):
  global rouge1, rouge2, rougeL

  tweet = row['reduced_tweets_700']

  generatedSummary = pipe(tweet, **gen_kwargs)[0]["summary_text"]

  rougeScore = ROUGE.get_scores(generatedSummary, row['Auto_summary'])

  f_rouge1 = rougeScore[0]['rouge-1']['f']
  f_rouge2 = rougeScore[0]['rouge-2']['f']
  f_rougel = rougeScore[0]['rouge-l']['f']

  rouge1 += f_rouge1
  rouge2 += f_rouge2
  rougeL += f_rougel

  return generatedSummary
```
